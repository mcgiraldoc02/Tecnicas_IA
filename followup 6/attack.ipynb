```python
from PIL import Image
import numpy as np
```

```python
# Cargar la imagen
imagen_path = '/content/drive/MyDrive/Técnicas_IA/followup 6/mangifera.jpg'  # Reemplaza con el nombre de tu imagen
imagen = Image.open(imagen_path)
```

```python
# Convertir la imagen a un arreglo de numpy
imagen_array = np.array(imagen)

# Aplicar la fórmula: imagen + 0.005 * x
resultado_array = imagen_array + 0.005 * imagen_array

# Asegurarse de que los valores estén dentro del rango de 0 a 255
resultado_array = np.clip(resultado_array, 0, 255)

# Convertir de nuevo a una imagen
resultado_imagen = Image.fromarray(resultado_array.astype('uint8'))

# Guardar la imagen resultante
resultado_imagen.save('/content/drive/MyDrive/Técnicas_IA/followup 6/mangifera_attack.jpg')


print("Imagen procesada y guardada ")
```
